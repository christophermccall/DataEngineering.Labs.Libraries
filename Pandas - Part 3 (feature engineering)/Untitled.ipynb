```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Combined Wine Data.csv')
```

```python
data.head()
```

```python
data['sweet or not'] = np.where(data['residual sugar'] >= data['residual sugar'].mean(), True, False) 
```

```python
data.head()
```

```python
quality = 6 
sweet = True
alcohol = data.alcohol.quantile(0.75)
color = 'red'
```

```python
condition = np.logical_and(data['quality'] >= quality, data['sweet or not'] == sweet)
condition = np.logical_and(condition, data['alcohol'] > alcohol)
#condition = np.logical_and(condition, data['color']> color)
#what does he mean by this ^^^ why are we searching for wines that are "greater" than 'red'
condition1 = np.logical_and(condition, data['color'] == 'white')
condition = np.logical_and(condition, data['color'] == color)
```

```python
condition.value_counts()
```

```python
data['good reds'] = condition
data['good whites'] = condition1
```

```python
data.head()
```

```python
data.loc[data['good reds'] == True]
```

```python
data.loc[data['good whites']== True]
```

```python
reds_i_like = data[data['good reds'] == True]
whites_i_like = data[data['good whites'] == True]
```

```python
reds_i_like.to_csv('my favorite red wines')
whites_i_like.to_csv('my favorite white wines')
```
